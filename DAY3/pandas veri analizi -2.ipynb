```python
# Yaşa göre hayatta kalma dağılım
plt.figure(figsize=(7,7))
sns.histplot(data=df,
            x='Age', hue='Survived',
            multiple='stack',
            kde=True,palette='pastel', bins=10, alpha=.5)
plt.title('Yaş - Hayatta Kalma Grafiği')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grafiği çiz
sns.barplot(x='Sex', y='Survived', data=df,ci=None, palette=['pink', 'blue'])

# Yüzdelik etiketleri ekle
for container in plt.gca().containers:
    #plt.bar_label(p, fmt='%1.2f%%', label_type='edge',rotation=90)
    #plt.bar_label(p, fmt='%1.2f%%')
    labels = [f'{int(value.get_height() * 100)}%' for value in container]
    plt.bar_label(container, labels=labels)
# Eksen etiketleri ve başlık
plt.xlabel('Cinsiyet')
plt.ylabel('Hayatta Kalma Oranı')
plt.title('Cinsiyete Göre Hayatta Kalma Durumu')

# Grafiği göster
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grafiği çiz
sns.barplot(x='Sex', y='Survived',hue='Pclass', data=df,ci=None, palette=['pink', 'blue', 'orange'])

# Yüzdelik etiketleri ekle
for container in plt.gca().containers:
    #plt.bar_label(p, fmt='%1.2f%%', label_type='edge',rotation=90)
    #plt.bar_label(p, fmt='%1.2f%%')
    labels = [f'{int(value.get_height() * 100)}%' for value in container]
    plt.bar_label(container, labels=labels)
# Eksen etiketleri ve başlık
plt.xlabel('Cinsiyet')
plt.ylabel('Hayatta Kalma Oranı')
plt.title('Sınıf ve Cinsiyete Göre Hayatta Kalma Durumu')

# Grafiği göster
plt.show()
```
