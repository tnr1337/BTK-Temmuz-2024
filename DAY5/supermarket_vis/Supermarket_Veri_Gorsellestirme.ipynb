## Veri Görselleştirme

```python
# Aylara Göre alışveriş 
plt.figure(figsize=(12,6)) # grafiğin boyutunu ayarlama
sns.countplot(x=df['month_name'])
plt.title('Aylara göre alışveriş sayısı')
plt.xticks(rotation=90) # x etiketi açısı
plt.show();
```

```python
# aylık satış sayılarını Şubelere göre grafiğini çizelim.
plt.figure(figsize=(12,6))
plt.title('Şubelerin aylara göre alışveriş sayısı')
sns.countplot(x=df['month_name'], hue=df['Branch']);
```

```python
# Cinsiyete göre aylık alışveriş sayıları
plt.title('Cinsiyete göre aylık alışveriş sayıları')
sns.countplot(x=df['month_name'],
             hue=df['Gender'])
```

```python
# Yatay
plt.title('Cinsiyete göre aylık alışveriş sayıları')
sns.countplot(y=df['month_name'],
             hue=df['Gender'])
plt.ylabel('months')
```

```python
#günlere göre alışveriş sayısı
sns.countplot(x=df['day_name'])
plt.xticks(rotation=45);
```

```python
# En yoğun saatler
sns.countplot(x=df['hour'])
plt.xticks(rotation=45);
```

```python
# En çok harcama yapılan saat - Saatlik toplam harcama 
df.groupby('hour')['Total'].sum().plot();
```

```python
# Ürün türlerine göre pasta grafiği
df['Product line'].value_counts().plot.pie(autopct='%1.1f%%',  #%1.1f%% - Format -> f%%  automatic percentage" (otomatik yüzde) 
                                           ylabel='',
                                           explode=[.5,0,0,0,0,0]); # 0.1'ler dilimler arasındaki açıklığı ifade eder.
```

```python
# Cinsiyete göre Ürün Türü Alışveriş Davranışlarının grafiği 
renkler = {'Male':'green','Female':'purple'}
sns.countplot(x=df['Product line'], hue=df['Gender'], palette = renkler)
plt.title('Cinsiyete göre Ürün Türü Alışveriş Davranışları')
plt.xticks(rotation=50);
```

```python
# Şubelere göre ürün türünün satış miktarları
plt.figure(figsize=(16,12))
sns.countplot(x=df['Product line'],
             hue=df['Branch']) 
plt.xticks(rotation=45)
plt.savefig('Şube_urun.png', dpi=300) #grafiği kaydetme
```

```python
#Her bir şubenin saatlik ciro - yapılan harcama tutarı 
df.groupby(['Branch', 'hour'])['Total'].sum().unstack(level='hour').plot(kind='bar', figsize=(12, 6));
```
